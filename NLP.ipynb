## Data cleaning


```python
un_emotions = df['emotions'].unique()
i = 0
emotions_dict = {}
for emotion in un_emotions:
    emotions_dict[emotion] = i
    i += 1
emotions_dict
```

```python
df['emotions'].apply(lambda x: emotions_dict[x])
```

```python
df['text'] = df['text'].apply(lambda x: x.lower())
df
```

```python
import string
```

```python
def remove_punctuation(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
df['text']=df['text'].apply(remove_punctuation)
```

```python
def remove_number(text):
  new = ''
  for i in text:
    if not i.isdigit():
      new += i
  return new
df['text']=df['text'].apply(remove_number)
```

```python
def remove_emoji(txt):
  new = ''
  for i in txt:
    if i.isascii():
      new +=i
  return new
df['text']=df['text'].apply(remove_emoji)
```

```python
import nltk
nltk.download
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt') #for tokenazatiob
```

```python
stop_words = set(stopwords.words('english'))
```

```python
len(stop_words)
```

```python
def remove(txt):
  words = txt.split()
  new_words = []
  for word in words:
    if word not in stop_words:
      new_words.append(word)
  return ' '.join(new_words)
df['text']=df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotions'], test_size=0.2, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer

bow_vect = CountVectorizer()
X_train_bow = bow_vect.fit_transform(X_train)
X_test_bow = bow_vect.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)
y_pred = nb_model.predict(X_test_bow)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#tf-idf with nayve basyes
tf_idf_vect = TfidfVectorizer()
X_train_tf_idf = tf_idf_vect.fit_transform(X_train)
X_test_tf_idf = tf_idf_vect.transform(X_test)
nb2_model = MultinomialNB()
```

```python
nb2_model.fit(X_train_tf_idf, y_train)
y_pred = nb2_model.predict(X_test_tf_idf)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#now tfidf with logistic regression
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
lr_model.fit(X_train_tf_idf, y_train)
y_pred = lr_model.predict(X_test_tf_idf)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
